```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as pl

%matplotlib inline
```

```python
dataframe = pd.read_csv(r"https://github.com/jseabold/depy/raw/master/data/adult.data.cleaned.csv.gz", compression="gzip")
```

```python
# Trasforming categorical features to numeric as seaborn can only processes numeric values
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
dataframe_num = dataframe.apply(LabelEncoder().fit_transform)
```

```python
# Scatter Metrics
#pd.scatter_matrix(dataframe_num, alpha = 0.3, figsize = (14,8), diagonal = 'kde');

pd.plotting.scatter_matrix(dataframe_num, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
```

```python
# Pairplot
sns.pairplot(dataframe_num);
```

```python
# Heatmap
import seaborn as sns

f, ax = pl.subplots(figsize=(10, 8))
corr = dataframe_num.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, annot=False);
```
